```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.spatial.distance import cosine
from pathlib import Path

MARS_DIR    = '/content/drive/MyDrive/Elevation New/mars'
ICELAND_DIR = '/content/drive/MyDrive/Elevation New/iceland'

# ── Load CSVs ─────────────────────────────────────────────────────────
def load_profiles(folder):
    profiles = []
    for f in sorted(Path(folder).glob('*.csv')):
        df = pd.read_csv(f)
        # Handle whatever column names QGIS used
        df.columns = [c.strip().lower() for c in df.columns]
        dist_col = [c for c in df.columns if 'dist' in c][0]
        elev_col = [c for c in df.columns if 'elev' in c or 'z' in c][0]
        dist = df[dist_col].values.astype(np.float32)
        elev = df[elev_col].values.astype(np.float32)
        profiles.append({'name': f.stem, 'dist': dist, 'elev': elev})
        print(f'  Loaded: {f.stem}  ({len(dist)} points, '
              f'{dist[-1]:.0f}m long, '
              f'elev range {elev.min():.1f}→{elev.max():.1f}m)')
    return profiles

print('Mars profiles:')
mars_profiles = load_profiles(MARS_DIR)
print(f'\nIceland profiles:')
iceland_profiles = load_profiles(ICELAND_DIR)
print(f'\n✅ Mars: {len(mars_profiles)}  Iceland: {len(iceland_profiles)}')
```

```python
# ── Compute metrics per profile ───────────────────────────────────────
def cone_metrics(dist, elev, label=''):
    elev = np.array(elev, dtype=np.float32)
    dist = np.array(dist, dtype=np.float32)

    # Normalize distance to start from 0
    dist = dist - dist[0]

    base_elev   = float(np.mean([elev[0], elev[-1]]))
    summit_elev = float(np.max(elev))
    summit_idx  = int(np.argmax(elev))
    center_elev = float(elev[len(elev)//2])
    cone_height = summit_elev - base_elev
    crater_depth = summit_elev - center_elev
    width_m     = float(dist[-1])
    h_w_ratio   = cone_height / width_m if width_m > 0 else 0

    peaks, props = find_peaks(elev, prominence=1.0)
    valleys, _   = find_peaks(-elev, prominence=1.0)

    return {
        'label'         : label,
        'base_elev_m'   : round(base_elev, 2),
        'summit_elev_m' : round(summit_elev, 2),
        'cone_height_m' : round(cone_height, 2),
        'crater_depth_m': round(crater_depth, 2),
        'width_m'       : round(width_m, 2),
        'h_w_ratio'     : round(h_w_ratio, 4),
        'num_peaks'     : len(peaks),
    }

mars_metrics    = [cone_metrics(p['dist'], p['elev'], f"Mars_{p['name']}")
                   for p in mars_profiles]
iceland_metrics = [cone_metrics(p['dist'], p['elev'], f"Iceland_{p['name']}")
                   for p in iceland_profiles]

all_metrics = pd.DataFrame(mars_metrics + iceland_metrics)
print('=== All Cone Metrics ===')
print(all_metrics.to_string(index=False))

# ── Domain comparison ─────────────────────────────────────────────────
mars_df    = all_metrics[all_metrics['label'].str.startswith('Mars')]
iceland_df = all_metrics[all_metrics['label'].str.startswith('Iceland')]

cols = ['cone_height_m', 'crater_depth_m', 'width_m', 'h_w_ratio']
comparison = pd.DataFrame({
    'Metric'  : cols,
    'Mars'    : [round(float(mars_df[c].mean()),    2) for c in cols],
    'Iceland' : [round(float(iceland_df[c].mean()), 2) for c in cols],
})
comparison['Ratio_Mars_Iceland'] = (
    comparison['Mars'] / comparison['Iceland']).round(3)

print('\n=== Mars vs Iceland Mean Comparison ===')
print(comparison.to_string(index=False))

# ── Shape similarity ──────────────────────────────────────────────────
def normalize(p):
    p = np.array(p, dtype=np.float32)
    rng = p.max() - p.min()
    return (p - p.min()) / rng if rng > 0 else np.zeros_like(p)

x = np.linspace(0, 1, 200)
sims = []
for m in mars_profiles:
    for ic in iceland_profiles:
        m_r  = np.interp(x, np.linspace(0,1,len(m['elev'])),  m['elev'])
        ic_r = np.interp(x, np.linspace(0,1,len(ic['elev'])), ic['elev'])
        sim  = 1 - cosine(normalize(m_r), normalize(ic_r))
        sims.append({'mars': m['name'], 'iceland': ic['name'],
                     'cosine_sim': round(float(sim), 4)})

sim_df = pd.DataFrame(sims)
print(f'\n=== Profile Shape Similarity ===')
print(f'Mean : {sim_df["cosine_sim"].mean():.4f}')
print(f'Min  : {sim_df["cosine_sim"].min():.4f}')
print(f'Max  : {sim_df["cosine_sim"].max():.4f}')

# ── Plot normalized overlaid profiles ────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 5))
for p in mars_profiles:
    p_r = np.interp(x, np.linspace(0,1,len(p['elev'])), p['elev'])
    ax.plot(x, normalize(p_r), color='#c0392b', alpha=0.4, lw=1.2,
            label='Mars' if p == mars_profiles[0] else '_')

for p in iceland_profiles:
    p_r = np.interp(x, np.linspace(0,1,len(p['elev'])), p['elev'])
    ax.plot(x, normalize(p_r), color='#2980b9', alpha=0.4, lw=1.2,
            label='Iceland' if p == iceland_profiles[0] else '_')

# Mean lines
mars_stack = np.array([np.interp(x, np.linspace(0,1,len(p['elev'])),
                                  p['elev']) for p in mars_profiles])
iceland_stack = np.array([np.interp(x, np.linspace(0,1,len(p['elev'])),
                                     p['elev']) for p in iceland_profiles])

ax.plot(x, normalize(mars_stack.mean(axis=0)),
        color='#922b21', lw=2.5, label='Mars mean')
ax.plot(x, normalize(iceland_stack.mean(axis=0)),
        color='#1a5276', lw=2.5, label='Iceland mean')

ax.set_xlabel('Normalised Distance along Profile')
ax.set_ylabel('Normalised Elevation')
ax.set_title('Mars vs Iceland — Normalised Cone Profile Comparison')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/Elevation New/profile_comparison.png',
            dpi=150, bbox_inches='tight')
plt.show()

# ── Save CSVs ─────────────────────────────────────────────────────────
all_metrics.to_csv('/content/drive/MyDrive/Elevation New/cone_metrics.csv',
                   index=False)
comparison.to_csv('/content/drive/MyDrive/Elevation New/comparison.csv',
                  index=False)
print('\n✅ All results saved')
```
